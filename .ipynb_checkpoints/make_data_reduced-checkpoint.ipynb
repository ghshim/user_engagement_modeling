```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./data/exhibition_behavior_preprocessed.csv', delimiter=',', index_col=False)
df.head()
```

```python
new_df = df.groupby('pid').agg({'code': list, 'A/C': 'first', 'M/F': 'first', 'duration': sum, 'engagement_score': 'first', 'norm_engagement_score': 'first', 'engagement_level': 'first'})
```
