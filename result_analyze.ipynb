```python
import pandas as pd
import numpy as np
```

```python
file_path = './data/final_behavior_hmm_result_n_states3.xlsx'

# Excel 파일 불러오기
xls = pd.ExcelFile(file_path)

# 시트 이름 확인
sheet_names = xls.sheet_names
print(sheet_names)  # 시트 이름 출력

# 각 시트를 데이터프레임으로 저장
dfs = {}  # 데이터프레임을 저장할 딕셔너리

for sheet_name in sheet_names:
    df = xls.parse(sheet_name)  # 시트 불러오기
    dfs[sheet_name] = df  # 데이터프레임을 딕셔너리에 저장
```

```python
dfs['state2']
```

```python
hidden_states = {'curious':dfs['state1'], 'attentive':dfs['state0'], 'engaged':dfs['state2']}
```

```python
# 각 데이터프레임의 'A/C'와 'M/F' 열의 통계 출력
for sheet_name, df in hidden_states.items():
    print(f"Statistics for sheet '{sheet_name}':")
    print(df[['A/C', 'M/F']].describe())
    print()
```

```python
print("-----------------------")
print("        curious        ")
print("-----------------------")
print(hidden_states['curious']['A/C'].value_counts())
print()
print(hidden_states['curious']['M/F'].value_counts())
print()
print("Total:", len(hidden_states['curious']))
```

```python
print("-----------------------")
print("        curious        ")
print("-----------------------")
child_male_count = hidden_states['curious'].loc[(hidden_states['curious']['A/C'] == 'Child') & (hidden_states['curious']['M/F'] == 'Male')].shape[0]
child_female_count = hidden_states['curious'].loc[(hidden_states['curious']['A/C'] == 'Child') & (hidden_states['curious']['M/F'] == 'Female')].shape[0]
adult_male_count = hidden_states['curious'].loc[(hidden_states['curious']['A/C'] == 'Adult') & (hidden_states['curious']['M/F'] == 'Male')].shape[0]
adult_female_count = hidden_states['curious'].loc[(hidden_states['curious']['A/C'] == 'Adult') & (hidden_states['curious']['M/F'] == 'Female')].shape[0]

print("Child - Male:", child_male_count)
print("Child - Female:", child_female_count)
print("Adult - Male:", adult_male_count)
print("Adult - Female:", adult_female_count)
```
